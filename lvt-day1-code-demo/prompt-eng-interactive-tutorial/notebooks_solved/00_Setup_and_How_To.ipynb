# Setup & How To

This course teaches prompt engineering using the Claude API in JavaScript. Each chapter is a Deno notebook that you run cell by cell, experimenting with prompts and inspecting model output directly. You'll need an Anthropic API key, which you can get at [console.anthropic.com](https://console.anthropic.com).

## Prerequisites & setup

1. **Install Deno** from [deno.land](https://deno.land). Deno is the runtime for all notebooks; no `npm install` required.
2. **Install the Deno Jupyter kernel**:
   ```
   deno jupyter --install
   ```
3. **Configure your API key**: copy `.env.example` to `.env` and set `ANTHROPIC_API_KEY`.
4. **Open notebooks** in VS Code (with the Jupyter extension) or `jupyter lab`, then select the **Deno** kernel for each notebook.

## The helper

All chapters import from `lib/helpers.js`, which wraps the Anthropic Messages API in a single `getCompletion(prompt, system?)` function and exports the pinned `MODEL` string.

The Messages API requires three things: `model`, `max_tokens`, and a `messages` array. The `system` prompt and `temperature` are optional. We pin `temperature: 0` throughout so results are deterministic, so the same prompt always produces the same output, which makes exercises predictable and lets you compare outputs across edits without randomness masking changes.

````typescript
import { getCompletion, MODEL } from "../lib/helpers.js";
console.log("Using model:", MODEL);
````

````typescript
console.log(await getCompletion("In one sentence, what is an idempotent HTTP method?"));
````

`getCompletion(prompt, system?)` is the only helper you need for most chapters. It is a thin wrapper around `client.messages.create` defined in [`lib/helpers.js`](../lib/helpers.js). The optional `system` argument defaults to an empty string (no system prompt).

Note that cells use top level `await`, which works in Deno notebook cells without any wrapper function.
